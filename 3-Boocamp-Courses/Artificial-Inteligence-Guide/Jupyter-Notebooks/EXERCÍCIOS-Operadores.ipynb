```python
print('Exercicio Qual é a Idade')
idade = int(input("Entre com a idade"))
if (idade >= 0) and (idade <= 12):
  print('É uma criança entre 0 e 12 anos')
else:
    if (idade >= 13) and (idade <= 17):
        print('É um adolescente entre 13 e 17 anos')
    else:
        if (idade >= 18):
          print('É um adulto acima de 18 anos')
        else:
            print('Idade inválida')
```
