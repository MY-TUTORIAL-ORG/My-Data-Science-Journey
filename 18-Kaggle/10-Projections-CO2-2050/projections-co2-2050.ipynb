```python
#word cloud
from wordcloud import WordCloud, ImageColorGenerator
text = " ".join(str(each) for each in df.variable)
# Create and generate a word cloud image:
wordcloud = WordCloud(max_words=200,colormap='Set3', background_color="black").generate(text)
plt.figure(figsize=(10,6))
plt.figure(figsize=(15,10))
# Display the generated image:
plt.imshow(wordcloud, interpolation='Bilinear')
plt.axis("off")
plt.figure(1,figsize=(12, 12))
plt.show()
```

```python
fig = px.bar(df, 
             x='year', y='unit', color_discrete_sequence=['#DC143C'],
             title='Units Projections till 2050', text='sector')

fig.update_layout(barmode='stack', xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
ax = df['sector'].value_counts()[:20].plot.barh(figsize=(12, 6), color='orange')
ax.set_title('Sectors', size=18, color='green')
ax.set_ylabel('Sector', size=10)
ax.set_xlabel('Count', size=10);
```

```python
ax = df['indicator'].value_counts()[:20].plot.barh(figsize=(12, 6), color='green')
ax.set_title('Indicators', size=18, color='orange')
ax.set_ylabel('Indicators', size=10)
ax.set_xlabel('Count', size=10);
```

```python
ax = df['variable'].value_counts()[:20].plot.barh(figsize=(12, 6), color='red')
ax.set_title('Variables', size=18, color='red')
ax.set_ylabel('Variables', size=10)
ax.set_xlabel('Count', size=10);
```

```python
ax = df['unit'].value_counts()[:20].plot.barh(figsize=(12, 6), color='purple')
ax.set_title('Units', size=18, color='red')
ax.set_ylabel('Units', size=10)
ax.set_xlabel('Count', size=10);
```
